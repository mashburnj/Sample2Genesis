```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from scipy.io import wavfile
import os
import numpy as np
```

```python
SampleName = "01 - [Prologue]_1.wav" # Audio File
os.chdir('./Training_data/')
SampleRate, Audiodata = wavfile.read(SampleName)

print('Audio data shape is')
print(np.shape(Audiodata))

# Plot the audio signal in time
plt.plot(Audiodata)
plt.title('Audio signal in time',size=16)
```

```python
# spectrum
from scipy.fft import fft # fourier transform
n = len(Audiodata) 
FFTSize = 1024
AudioFreq = fft(Audiodata,n=FFTSize,axis=0)
AudioFreq = AudioFreq[0:int(np.ceil((n+1)/2.0))] #Half of the spectrum
MagFreq = np.abs(AudioFreq) # Magnitude
MagFreq = MagFreq / float(n)

print('FFT of audio data shape is')
print(np.shape(MagFreq))
print(MagFreq.max())

plt.figure()
freqAxis = np.arange(0,int(np.ceil((n+1)/2.0)), int(np.ceil((n+1)/2.0))/FFTSize) * (SampleRate / n);
plt.plot(freqAxis/1000.0, MagFreq) #Spectrum
plt.xlabel('Frequency (kHz)'); plt.ylabel('Spectrum');
plt.xlim([0, 0.5])
```

```python
# basic smoothing/binning. original size is (17641,)

binned = MagFreq[:,0].copy()
numberInBin = 4
numberOfZerosToAdd = numberInBin - np.shape(MagFreq)[0] % numberInBin
binned = np.concatenate((binned,np.zeros(numberOfZerosToAdd))) # last bin will have a small avg, but it'll be small anyway.

smoothed = binned.reshape(np.shape(MagFreq)[0]//numberInBin + 1, numberInBin).mean(1)
print(np.shape(smoothed))

plt.figure()
freqAxisSmoothed = np.arange(0,int(np.ceil((n+1)/2.0)), numberInBin) * (SampleRate / n);
plt.plot(freqAxisSmoothed/1000.0, smoothed) #Smoothed spectrum
plt.xlabel('Frequency (kHz)'); plt.ylabel('Smoothed Spectrum');
plt.xlim([0, 0.5])
```

```python
# power spectrum
MagFreq = MagFreq**2
if n % 2 > 0: # ffte odd 
    MagFreq[1:len(MagFreq)] = MagFreq[1:len(MagFreq)] * 2
else:# fft even
    MagFreq[1:len(MagFreq) -1] = MagFreq[1:len(MagFreq) - 1] * 2 

plt.figure()
freqAxis = np.arange(0,int(np.ceil((n+1)/2.0)), 1.0) * (SampleRate / n);
plt.plot(freqAxis/1000.0, 10*np.log10(MagFreq)) #Power spectrum
plt.xlabel('Frequency (kHz)'); plt.ylabel('Power spectrum (dB)');
#plt.xlim([0, 0.5])
#plt.ylim([-30,-10])
```

```python
# smoothing with a Hann window
windowed = Audiodata.copy() #clone Audiodata
windowed[:,1] = Audiodata[:,1]*np.hanning(np.shape(Audiodata)[0])
AudioFreqSmooth = fft(windowed)
AudioFreqSmooth = AudioFreqSmooth[0:int(np.ceil((n+1)/2.0))] #Half of the spectrum
MagFreqSmooth = np.abs(AudioFreqSmooth) # Magnitude
MagFreqSmooth = MagFreqSmooth / float(n)

plt.figure()
freqAxis = np.arange(0,int(np.ceil((n+1)/2.0)), 1.0) * (SampleRate / n);
plt.plot(freqAxis/1000.0, MagFreqSmooth) #Spectrum
plt.xlabel('Frequency (kHz)'); plt.ylabel('Spectrum');
plt.xlim([0, 0.5])
```

```python
# basic smoothing/binning. original size is (17641,)

binnedSmooth = MagFreqSmooth[:,0].copy()
numberOfZerosToAdd = numberInBin - np.shape(MagFreqSmooth)[0] % numberInBin
binnedSmooth = np.concatenate((binned,np.zeros(numberOfZerosToAdd))) # last bin will have a small avg, but it'll be small anyway.

smoothedSmooth = binned.reshape(np.shape(MagFreq)[0]//numberInBin + 1, numberInBin).mean(1)
print(np.shape(smoothedSmooth))

plt.figure()
freqAxisSmoothedSmooth = np.arange(0,int(np.ceil((n+1)/2.0)), numberInBin) * (SampleRate / n);
plt.plot(freqAxisSmoothedSmooth/1000.0, smoothedSmooth) #Smoothed spectrum
plt.xlabel('Frequency (kHz)'); plt.ylabel('Smoothed Spectrum after Hanning');
plt.xlim([0, 0.5])
```
